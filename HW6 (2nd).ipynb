#  Задание 2

Даны два слова, ваша программа должна проверять является ли второе слово обратной записью первого слова.

```python
# первая пара слов

a = "dog"
b = "cat"
c = a[::-1]
if c == b:
  print("YES")
else:
  print("NO")
```

```python
# вторая пара слов
a = "live"
b = "evil"
c = a[::-1]
if c == b:
  print("YES")
else:
  print("NO")
```
